# Introduction
## Why loan Data?
This Data was selected because I think it is important to be transparent regarding money. It is good to have an algorithm that the loaner can use in order to determine if the individual should recieve the loan as well as for the individual to be able to know ahead of time if they would qualify for a loan given some factors.
## How do we solve it?
To solve this problem, we will use a supervised learning task and take advantage of machine learning algorithms such as logistic Regression, Decision Trees, K-nearest neighbors, and Support Vector Machines.
## Where did the data come from?
The data to make this project possible was provided as a playground competition on [https://www.kaggle.com/competitions/playground-series-s4e10/data](https://www.kaggle.com/competitions/playground-series-s4e10/data). This data from kaggle was generated from a model prediction on a loan approval dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import seaborn as sns
import warnings

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Data/loan data/train.csv")
display(data.head(5))
print("shape of data: {}".format(data.shape))
```

## Data Information
Our training data is provided in a csv file which we convert into a pandas.DataFrame object. It has 58,645 rows and 13 columns (12 features, 1 label) while also containing no null values. We see that there are 7 meaningful continuous values such as person_age, person_income, person_emp_length, etc. There are also 4 categorical values such as person_home_ownership, loan_intent, loan_grade, and cb_person_default_on_file. Finally, our prediction value will take a binary value of 0 for not approved and 1 for the loan being approved.

```python
# check data to see if we have to do imputations
print(data['person_home_ownership'].unique())
print(data['loan_intent'].unique())
print(data['loan_grade'].unique())
print(data['cb_person_default_on_file'].unique())

categoricalCols = ['person_home_ownership', 'loan_intent', 'loan_grade', 'cb_person_default_on_file']
continuousCols = ['person_age', 'person_income', 'person_emp_length', 'loan_amnt', 'loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length']
```


We can first perform a correlation test on data that we know is continuous and see if there are any values that stand out. We can see that there is a very strong correlation with person_age and cb_person_cred_hist_length, and a fairly strong correlation between loan_amnt and loan_percent_income.

# Data Cleaning
We will now transform the categorical data into something that we can pass into our supervised machine learning algorithms. In order to do this, ordinal categories will be given a numerical representation and nominal data will be one-hot encoded because of the small amount of unique categories

```python
from sklearn.preprocessing import OrdinalEncoder, LabelBinarizer, OneHotEncoder
homeOwnership = OrdinalEncoder(categories=[['OTHER', 'RENT', 'MORTGAGE','OWN']])
data['person_home_ownership'] = homeOwnership.fit_transform(data[['person_home_ownership']])
```

```python
# there is no order for loan amounts and since the dimentionality of the data is already fairly small, we will use One hot encoding
loanIntent = OneHotEncoder()
temp = loanIntent.fit_transform(data[['loan_intent']]).toarray()
encodedDF = pd.DataFrame(temp, columns = loanIntent.get_feature_names_out())
data = pd.concat([data, encodedDF], axis=1)
data.drop(columns=['loan_intent'], inplace=True)
```

```python
# loan_grade is risk associated with loan/person, Grade A is lowest risk and risk increases as you go down
# this is why we will use an ordinal encoder with A giving the highest score
loanGrade = OrdinalEncoder(categories=[['G', 'F', 'E','D', 'C', 'B', 'A']])
data['loan_grade'] = loanGrade.fit_transform(data[['loan_grade']])
```

```python
defaultOnFile = OrdinalEncoder(categories=[['Y', 'N']]) # Setting Y to 0 because it will not help you get approved for loan if on record
data['cb_person_default_on_file'] = defaultOnFile.fit_transform(data[['cb_person_default_on_file']])
display(data)
```

Now that the data is represented as numerical values and interpretable by algorithms, we can check some statistics on it

```python
display(data.describe())
```

We see things like person_emp_length which has a max value of 123. We are going to assume that a person has not been working for 123 years and remove data above 70 years working because these cases are less likely. Fortunately, there were only 2 cases of individuals working more than 70 years. There is another outlier in loan_percent_income, however we will leave it because it seems reasonable that a student with little to no income may as for a loan for school and such and have a high percentage

```python
data = data[data['person_emp_length'] < 70]
print(data.shape)
```

```python
display(data.describe())
```

Now that we have the data cleaned, we will check the correlation again and we can see that loan_grade and loan_int_rate are negatively correlated as well as a few others.

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data.corr(), annot=True, fmt='.1f')
```

# Data scaling
Now we are going to try to scale the data to follow a normal distribution to show how the accuracy changes. We expect an improvement in algorithms such as KNN because of how the distance metric is used. We will also create a seperate train and test split to see how our data performs with unseen data at the end. This is used because if KNN is trained on all the data, it accurately predicts all the labels and we can not tell any measure metrics from it in this case. K-fold cross validation is still used to select paramaters for each algorithm.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
Xscaled = Xdata.copy()
Xscaled[continuousCols] = StandardScaler().fit_transform(Xscaled[continuousCols])
display(Xscaled.head(5))
Xscaled_train, Xscaled_val, y_train, y_val = train_test_split(Xscaled, Ydata, test_size=.2, random_state=42)
```

lets check to make sure there are about the same distribution of 1's and 0's in the train and validation data

```python
display(y_train.value_counts())
display(y_val.value_counts())
```

# Logistic Regression
Lets try refitting our logistic regression model with scaled data

```python
logRegModel = LogisticRegression()
grid = {'penalty': ['None', 'l1', 'l2', 'elasticnet'], 
        'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
       'C': list(range(1,5)),
       'random_state': [42]}
search = GridSearchCV(logRegModel, grid, cv = 5).fit(Xscaled_train, y_train)
```

```python
print(search.best_score_)
print(search.best_params_)
```

<div class="alert alert-block alert-warning">
As We can see, the accuracy for the logisticRegression model did not improve and still has a score of about 90%

# KNN
Now lets try training our KNN model on the scaled training data and doing the same paramater search from before we scaled

```python
neighborClassifier = KNeighborsClassifier()
neighborGrid = {'n_neighbors': list(range(1,10)),
               'weights': ['uniform', 'distance'],
               'p': [1,2]
               }
neighborSearchScaled = GridSearchCV(neighborClassifier, neighborGrid, cv = 3).fit(Xscaled_train, y_train) # add verbose=10 to constructor for logs
```

```python
print(neighborSearchScaled.best_score_)
print(neighborSearchScaled.best_params_)
```

Since our best value for n_neighbors was at our max value in our grid, we will increase the search for best n_neighbors

```python
neighborGrid2 = {'n_neighbors': list(range(10,60, 10)),
               'weights': ['uniform', 'distance'],
               'p': [1,2]
               }
neighborSearchScaled2 = GridSearchCV(neighborClassifier, neighborGrid2, cv = 3).fit(Xscaled_train, y_train) # add verbose = 10 for logs
```

```python
print(neighborSearchScaled2.best_score_)
print(neighborSearchScaled2.best_params_)
```

<div class="alert alert-block alert-success">

Our KNN model fit with the scaled data gets an accuracy of about 93.5% which is an improvement over both our logistic regression and earlier KNN model.

# XGBClassifier
We will try to fit an extreme Gradient boost classifier to the data and see how well it performs. Extreme gradient boosting trees have become popular lately and have a tendency to be very fast and not overfit so they are well suited for many different problems.

```python
from xgboost import XGBClassifier
xgbModel = XGBClassifier(objective = 'binary:logistic', random_state=42)
xgbGrid = {'n_estimators': list(range(50, 350, 50)),
          'max_depth': list(range(1,3)),
          'colsample_bytree' : [x / 10 for x in range(0,10)]}
xgbSearch = GridSearchCV(xgbModel, xgbGrid, cv=5).fit(Xscaled_train, y_train) # set verbose = 10 in constructor
```

```python
print(xgbSearch.best_score_)
print(xgbSearch.best_params_)
```

Because our best score was at the maximums for both of our ranges, we will rerun with them increased

```python
xgbGrid = {'n_estimators': list(range(300, 600, 50)),
          'max_depth': list(range(1,7)),
          'colsample_bytree' : [x / 10 for x in range(0,10)]}
xgbSearch2 = GridSearchCV(xgbModel, xgbGrid, cv=5).fit(Xscaled, Ydata) # set verbose=10 in constructor
```

```python
print(xgbSearch2.best_score_)
print(xgbSearch2.best_params_)
# We can see that the best number of estimators was 300 with a max depth of 3 and less than half of the features samples on each tree
#, this is similar to the result we saw above with a max depth of 2.
```

<div class="alert alert-block alert-success">

As we can see from our grid search on XGBTrees, we get an accuracy of about 95% which is better than any other model we have currently fit

# Results and Analysis
The 2 metrics we are using to analyze the models are classification accuracy as well as AUC. To view classification accuracy past the values we obtained earlier, we will plot a confusion matrix on the validation set from our scaled data.

## Confusion Matrices

```python
from sklearn.metrics import confusion_matrix, roc_auc_score, ConfusionMatrixDisplay, RocCurveDisplay
xgbEstimator = xgbSearch.best_estimator_
disp = ConfusionMatrixDisplay.from_estimator(xgbEstimator, Xscaled_val, y_val, values_format="d")
plt.close()
fig, ax = plt.subplots()
disp.plot(ax=ax)
ax.set_title('XGB Model Confusion Matrix')
plt.show()
```

```python
knnEstimator = neighborSearchScaled.best_estimator_
disp = ConfusionMatrixDisplay.from_estimator(knnEstimator, Xscaled_val, y_val, values_format="d")
plt.close()
fig, ax = plt.subplots()
disp.plot(ax=ax)
ax.set_title('KNN Model Confusion Matrix')
plt.show()
```

```python
disp = ConfusionMatrixDisplay.from_estimator(svcModel, Xscaled_val, y_val, values_format="d")
plt.close()
fig, ax = plt.subplots()
disp.plot(ax=ax)
ax.set_title('SVC Model Confusion Matrix')
plt.show()
```

We can see from the confusion matrices that there are a lot more 0's than there are 1's and it is more common for all of our models to inaccurately predict a 1 as a 0 rather than a 0 as a 1. Our best model in this case is XGB because it predicts the most approved values correctly and gets almost the same number of denied values correct.

## ROC Curves

```python
disp = RocCurveDisplay.from_estimator(xgbEstimator, Xscaled_val, y_val)
plt.close()
fig, ax = plt.subplots()
disp.plot(ax=ax)
ax.set_title('XGB Model ROC Curve')
plt.show()
```

```python
disp = RocCurveDisplay.from_estimator(knnEstimator, Xscaled_val, y_val)
plt.close()
fig, ax = plt.subplots()
disp.plot(ax=ax)
ax.set_title('KNN Model ROC Curve')
plt.show()
```

```python
disp = RocCurveDisplay.from_estimator(svcModel, Xscaled_val, y_val)
plt.close()
fig, ax = plt.subplots()
disp.plot(ax=ax)
ax.set_title('SVC Model ROC Curve')
plt.show()
```

As we suspected from the earlier visualizations and scores, the XGB Model has the highest AUC which also means it is the best at seperating the classes properly. KNN is the second best and SVC is the worst of our good models.
